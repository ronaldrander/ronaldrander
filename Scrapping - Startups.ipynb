```python
import requests
from bs4 import BeautifulSoup
import sqlite3
import pandas as pd
import numpy as np
from selenium import webdriver
from time import sleep
```

```python
conn = sqlite3.connect('Startups.db')
```

```python
cur = conn.cursor()
```

```python
cur.execute('CREATE TABLE IF NOT EXISTS startups(nomes TEXT, segmentos TEXT, localizações TEXT, descrições TEXT, publicoalvos TEXT, receitas TEXT, momentos TEXT, sobres TEXT, anos TEXT, times TEXT, sites TEXT, logos TEXT, socios1 TEXT, fotos1 TEXT, cargos1 TEXT, linkedins1 TEXT, socios2 TEXT, fotos2 TEXT, cargos2 TEXT, linkedins2 TEXT, redessocial1 TEXT, redessocial2 TEXT, redessocial3 TEXT, redessocial4 TEXT, aceleradoras1 TEXT, aceleradoras2 TEXT, aceleradoras3 TEXT, aceleradoras4 TEXT)')
conn.commit()
```

```python
driver = webdriver.Firefox()

driver.get('https://startupbase.com.br/home/startups?q=&states=all&cities=all&segments=E-commerce~Agroneg%C3%B3cio&targets=all&phases=all&models=all&badges=all')
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
soup
```

```python
name = soup.find(class_='ais-InfiniteHits').find(class_='search-body__item')

for i in name.find_next_siblings():
    
    cur.execute('SELECT nomes FROM startups')
    empresas = [empresa[0] for empresa in cur.fetchall()]

    if i.find(class_='org__title sb-size-6').get_text().strip() in empresas:
        continue
    else:
        
        try:
            nome = i.find(class_='org__title sb-size-6').get_text().strip()
        except:
            nome = 'None'
            pass
    
        try:
            localização = i.find(class_='org__label').find('span').get_text().strip()
        except:
            localização = 'None'
            pass
    
        try:
            segmento = i.find(class_='org__label').find('span').find_next_sibling().get_text().strip()
        except:
            segmento = i.find(class_='org__label').find('span').get_text().strip()
            localização = 'None'
            pass
    
        try:
            descrição = i.find(class_='org__body').find(class_='org__description sb-size-10').get_text().strip()
        except:
            descrição = 'None'
            pass
    
        try:
            logo = i.find(class_='is-radius1').get('src')
        except:
            logo = 'None'
            pass
        
        link = 'https://startupbase.com.br' + i.find('a').get('href')
        driver2 = webdriver.Firefox()
        driver2.get(link)
        sleep(9)
    
        soup2 = BeautifulSoup(driver2.page_source, 'html.parser')
    
        try:
            publicoalvo = soup2.find(class_='startup-timely__info').find_next_sibling().find(class_='startup-timely__data has-text-weight-semibold sb-size-6').get_text().strip()
        except:
            publicoalvo = 'None'
            pass
    
        try:
            receita = soup2.find(class_='startup-timely__info').find_next_sibling().find_next_sibling().find(class_='startup-timely__data has-text-weight-semibold sb-size-6').get_text().strip()
        except:
            receita = 'None'
            pass
    
        try:
            momento = soup2.find(class_='startup-timely__info').find_next_sibling().find_next_sibling().find_next_sibling().find(class_='startup-timely__data has-text-weight-semibold sb-size-6').get_text().strip()
        except:
            momento = 'None'
            pass
        
        try:
            sobre = soup2.find(class_='publ-text sb-size-9').get_text().strip()
        except:
            sobre = 'None'
            pass
        
        try:
            ano = soup2.find(class_='publ-aside__info').find_next_sibling().find_next_sibling().find(class_='publ-text sb-size-10').get_text().strip()
        except:
            ano = 'None'
            pass
        
        try:
            time = soup2.find(class_='publ-aside__info').find_next_sibling().find_next_sibling().find_next_sibling().find(class_='publ-text sb-size-10').get_text().strip()
        except:
            time = 'None'
            pass
    
        try:
            site = soup2.find(class_='publ-channel is-radius3 is-unselectable').get('href')
        except:
            site = 'None'
            pass
    
        try:
            founders = soup2.find(class_='group-body startup-founders')
        except:
            pass
    
        try:
            socio1 = founders.find(class_='column is-half-touch is-one-quarter-desktop').find(class_='member__name sb-size-10').get_text()
        except:
            socio1 ='None'
            pass
    
        try:
            foto1 = founders.find(class_='column is-half-touch is-one-quarter-desktop').find(class_='is-rounded').get('src')
        except:
            foto1 = 'None'
            pass
    
        try:
            cargo1 = founders.find(class_='column is-half-touch is-one-quarter-desktop').find(class_='member__title').get_text()
        except:
            cargo1 = 'None'
            pass
    
        try:
            linkedin1 = founders.find(class_='column is-half-touch is-one-quarter-desktop').find(class_='member__link').get('href')
        except:
            linkedin1 = 'None'
            pass
    
        try:
            socio2 = founders.find(class_='column is-half-touch is-one-quarter-desktop').find_next_sibling().find(class_='member__name sb-size-10').get_text()
        except:
            socio2 = 'None'
            pass
    
        try:
            foto2 = founders.find(class_='column is-half-touch is-one-quarter-desktop').find_next_sibling().find(class_='is-rounded').get('src')
        except:
            foto2 = 'None'
            pass
    
        try:
            cargo2 = founders.find(class_='column is-half-touch is-one-quarter-desktop').find_next_sibling().find(class_='member__title').get_text()
        except:
            cargo2 = 'None'
            pass
    
        try:
            linkedin2 = founders.find(class_='column is-half-touch is-one-quarter-desktop').find_next_sibling().find(class_='member__link').get('href')
        except:
            linkedin2 = 'None'
            pass
    
        try:
            redes_sociais = soup2.find(class_='publ-social')
        except:
            pass
    
        try:
            redesocial1 = redes_sociais.find(class_='publ-social__link has-text-centered').get('href')
        except:
            redesocial1 = 'None'
            pass
    
        try:
            redesocial2 = redes_sociais.find(class_='publ-social__link has-text-centered').find_next_sibling().get('href')
        except:
            redesocial2 = 'None'
            pass
    
        try:
            redesocial3 = redes_sociais.find(class_='publ-social__link has-text-centered').find_next_sibling().find_next_sibling().get('href')
        except:
            redesocial3 = 'None'
            pass
        
        try:
            redesocial4 = redes_sociais.find(class_='publ-social__link has-text-centered').find_next_sibling().find_next_sibling().find_next_sibling().get('href')
        except:
            redesocial4 = 'None'
            pass
        
        try:
            aceleradoras = soup2.find(class_='columns is-mobile is-multiline')
        except:
            pass

        try:
            aceleradora1 = aceleradoras.find(class_='column is-half-mobile is-one-third-tablet is-one-quarter-desktop').find(class_='badge__name sb-size-10 has-text-weight-bold').get_text()
        except:
            aceleradora1 = 'None'
            pass
    
        try:
            aceleradora2 = aceleradoras.find(class_='column is-half-mobile is-one-third-tablet is-one-quarter-desktop').find_next_sibling().find(class_='badge__name sb-size-10 has-text-weight-bold').get_text()
        except:
            aceleradora2 = 'None'
            pass
    
        try:
            aceleradora3 = aceleradoras.find(class_='column is-half-mobile is-one-third-tablet is-one-quarter-desktop').find_next_sibling().find_next_sibling().find(class_='badge__name sb-size-10 has-text-weight-bold').get_text()
        except:
            aceleradora3 = 'None'
            pass
        
        try:
            aceleradora4 = aceleradoras.find(class_='column is-half-mobile is-one-third-tablet is-one-quarter-desktop').find_next_sibling().find_next_sibling().find_next_sibling().find(class_='badge__name sb-size-10 has-text-weight-bold').get_text()
        except:
            aceleradora4 = 'None'
            pass
   
        cur.execute('INSERT INTO startups(nomes, segmentos, localizações, descrições, publicoalvos, receitas, momentos, sobres, anos, times, sites, logos, socios1, fotos1, cargos1, linkedins1, socios2, fotos2, cargos2, linkedins2, redessocial1, redessocial2, redessocial3, redessocial4, aceleradoras1, aceleradoras2, aceleradoras3, aceleradoras4) VALUES(?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)', (nome, segmento, localização, descrição, publicoalvo, receita, momento, sobre, ano, time, site, logo, socio1, foto1, cargo1, linkedin1, socio2, foto2, cargo2, linkedin2, redesocial1, redesocial2, redesocial3, redesocial4, aceleradora1, aceleradora2, aceleradora3, aceleradora4))
        conn.commit()
        driver2.close()
        sleep(2)
```
